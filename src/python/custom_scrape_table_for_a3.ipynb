```python
import pandas as pd
import os
from bs4 import BeautifulSoup
import sqlite3
```

```python
# Read the HTML file
data_dir = '/Users/jnaid/demo_test/data_wrangling_304/data/assignment_1/raw/' 
html_file = os.path.join(data_dir, 'hello_class_jay.html')
print(html_file)
with open(html_file, 'r') as html_file:
    soup = BeautifulSoup(html_file, 'html.parser')
```

```python
# Get the h1 tag content to use in the "name" column and as the CSV file name
h1_tag = soup.find('h1')
name_value = h1_tag.text.strip()
csv_file_name = '/Users/jnaid/demo_test/data_wrangling_304/data/assignment_1/altered/' + name_value + '.csv'
print(csv_file_name)
```

```python
# Find the table
table = soup.find('table')
print(table)
```

```python
# Extract table headers
headers = [th.text.strip() for th in table.find_all('th')]

# Add the "name" column to the headers
headers.append('Name')

# Extract table rows and add the "name" column value
rows = []
for row in table.find_all('tr')[1:]:  # Skipping the header row
    cols = [td.text.strip() for td in row.find_all('td')]
    cols.append(name_value)  # Add the name column value
    rows.append(cols)

print(rows)
```

```python
# Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers)

# Save the DataFrame to a CSV file
own_csv_file = name_value + '.csv'
df.to_csv(csv_file_name, index=False, encoding='utf-8')

print(f"Data extracted and saved to {csv_file_name}")
```

```python
def input_page_output_directory(submission_folder_path):
    dir_list = []
    for each_file in os.listdir(submission_folder_path):
        if '.html' in each_file:
            full_path = os.path.join(submission_folder_path, each_file)
            dir_list.append(full_path)

    return dir_list
            #data_dir = '/Users/jnaid/demo_test/data_wrangling_304/data/assignment_1/raw/' 
            #html_file = os.path.join(data_dir, 'hello_class_jay.html')
            
#with open(html_file, 'r') as html_file:
#    soup = BeautifulSoup(html_file, 'html.parser')
x = '/Users/jnaid/demo_test/data_wrangling_304/data/assignment_1/raw/downloaded_class_submissions/'
list_of_directories = input_page_output_directory(x)
list_of_directories
    
```

```python
def names_list_maker(dir_list):
    name_list = []
    for listed_file in dir_list:
        with open(listed_file, 'r') as el_file:
            soup1 = BeautifulSoup(el_file, 'html.parser')
        h1_tag1 = soup1.find('h1')
        name_value1 = h1_tag1.text.strip()
        name_list.append(name_value1)

    return name_list

list_of_names = names_list_maker(list_of_directories)
```
